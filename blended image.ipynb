```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img2 = cv2.imread('not.jpg')
img = cv2.imread('Toy-Poodle.jpg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(img2)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img.shape
```

```python
img2.shape
```

```python
#blend images of the same size
img = cv2.resize(img,(2000,1200))
img2 = cv2.resize(img2,(2000,1200))
```

```python
plt.imshow(img)
```

```python
plt.imshow(img2)
```

```python
blend = cv2.addWeighted(src1=img, alpha=0.5, src2=img2, beta=0.5, gamma=0)
```

```python
plt.imshow(blend)
```

```python
blend = cv2.addWeighted(src1=img, alpha=0.5, src2=img2, beta=0.1, gamma=0.4)
```

```python
plt.imshow(blend)
```
